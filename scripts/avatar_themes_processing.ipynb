```python
import pandas as pd
df = pd.read_csv("data/raw/avatar.csv", encoding='latin1')
df = df.dropna(subset=['full_text'])
```

```python
themes = {
    "balance": [
        "balance",
        "harmony",
        "four nations",
        "avatar cycle",
        "restore",
        "spirit",
        "avatar state",
        "four elements",
        "equal",
        "peace",
        "natural order",
        "spirit world",
        "light and dark",
        "yin and yang",
        "keep the balance",
        "world out of balance"
    ],

    "destiny": [
        "destiny",
        "fate",
        "choose",
        "choice",
        "path",
        "who you are",
        "make your own",
        "meant to be",
        "honor",
        "duty",
        "purpose",
        "train",
        "learn",
        "Iâm just a kid",
        "Iâm the Avatar",
        "my own person",
        "decide",
        "to master",
        "future",
        "become"
    ],

    "war": [
        "war",
        "fire nation",
        "attack",
        "battle",
        "fight",
        "invasion",
        "soldiers",
        "army",
        "Air Nomads",
        "refugees",
        "colonies",
        "Phoenix King",
        "suffering",
        "my people",
        "loss",
        "destroy",
        "capture",
        "enemy"
    ],

    "redemption": [
        "redeem",
        "redemption",
        "forgive",
        "forgiveness",
        "change",
        "second chance",
        "sorry",
        "honor",
        "weakness",
        "strength",
        "regret",
        "shame",
        "make amends",
        "past mistakes",
        "I was wrong",
        "I'm sorry",
        "find my way",
        "turn my back",
        "new path"
    ]
}
for theme, words in themes.items():
    df[theme] = df['full_text'].str.lower().str.count('|'.join(words))
```

```python
water = df[df['book'] == 'Water']
earth = df[df['book'] == 'Earth']
fire = df[df['book'] == 'Fire']
```

```python

timelines = {"water": {}, "earth": {}, "fire": {}}
book_dfs = {"water": water, "earth": earth, "fire": fire}
for b, book_df in book_dfs.items():
    timelines[b] = book_df.groupby('chapter_num')[list(themes)].sum()


water_episodes = pd.DataFrame(timelines["water"].reset_index())
earth_episodes = pd.DataFrame(timelines["earth"].reset_index())
fire_episodes = pd.DataFrame(timelines["fire"].reset_index())

full_dataset = pd.concat([water_episodes.assign(book='Water'),
                          earth_episodes.assign(book='Earth'), fire_episodes.assign(book='Fire')], ignore_index=True)

full_dataset.to_csv('data/processed/avatar_themes.csv', index=False)
```
